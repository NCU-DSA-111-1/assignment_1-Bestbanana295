```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

```python
## read output.txt 
df = pd.read_csv('output.txt', sep='\t', header=None)

x =np.linspace(1,800,800)
plt.plot(x,df)

```

```python
## plot data
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
fig, ax = plt.subplots(figsize=(10, 6))
ax = sns.lineplot(x=0, y=1, data=df, ci=None, color='black')
ax.set(xlabel='Iteration', ylabel='Cost')
ax.set_title('Cost Function')

plt.savefig('cost.png', dpi=300, bbox_inches='tight')
plt.show()
```
